# IPL DATA ANALYSIS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```


# MOST WINS IN IPL

```python
temp=pd.DataFrame({'Winner':data['winner']})
count_wins=temp.value_counts()
labels=[X[0] for X in count_wins.keys()]
bar,ax=plt.subplots(figsize=(25,15))
ax=plt.pie(x=count_wins,autopct='%.2f%%',labels=labels)
plt.title("MOST WINS IN IPL",fontsize=17)
plt.show()
```

# HIGHEST NUMBER OF WINS IN ELIMIMNATOR

```python
sns.countplot(data['winner'][data['eliminator']=='Y'],data=data)
plt.title('HIGHEST NUMBER OF WINS IN ELIMINATOR')
plt.xticks(rotation=90)
plt.show()
```

# TOSS RESULT ANALYSIS

```python
temp=pd.DataFrame({'Toss Results':data['toss_decision']})
count_pom=temp.value_counts()
labels=[X[0] for X in count_pom.keys()]
bar,ax=plt.subplots(figsize=(18,10))
ax=plt.pie(x=count_pom,autopct='%.0f%%',labels=labels)
plt.title("TOSS RESULTS PERCENTAGE",fontsize=17)
plt.show()
```

# NUMBER OF WINS BY TEAMS WHEN BATTING 2ND

```python
sns.countplot(data['winner'][data['result']=='wickets'],data=data)
plt.title('COUNT OF WINS BY THE TEAMS BATTING 2ND')
plt.xticks(rotation=90)
plt.show()
```

# FAMOUS VENUE

```python
sns.barplot(x=data['venue'].value_counts().head(10).values,
           y=data['venue'].value_counts().head(10).index)
plt.title('Famous Venue')
plt.xlabel('Venue Count')
plt.ylabel('Venue')
plt.show()
```

# TOP 10 PLAYER OF THE MATCH AWARDS

```python
sns.barplot(x=data['player_of_match'].value_counts().head(10).values,
           y=data['player_of_match'].value_counts().head(10).index)
plt.title('TOP 10 PLAYER OF MATCH')
plt.xlabel('NO.OF AWARDS')
plt.ylabel('PLAYER NAME')
plt.show()
```

```python
sns.barplot(x=data['umpire1'].value_counts().head(10).values,
           y=data['umpire1'].value_counts().head(10).index)
plt.title('TOP 10 UMPIRES')
plt.xlabel('Match Count')
plt.ylabel('Umpire-1')
plt.show()
```
