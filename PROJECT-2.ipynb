# TRAINING TESTING & FITTING OF DATASET

```python
X=customers[['Avg. Session Length', 'Time on App', 'Time on Website','Length of Membership']]
y=customers['Yearly Amount Spent']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.83, random_state=300)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
l=LinearRegression()
```

```python
l.fit(X_train,y_train)
```

```python
coeff=pd.DataFrame(l.coef_,X.columns,columns=['Coefficient'])
coeff
```

# PREDICTIONS

```python
predict_variable=l.predict(X_test)
```

# SCATTER PLOT FOR THE PREDICTIONS DONE

```python
plt.scatter(y_test,predict_variable)
```
